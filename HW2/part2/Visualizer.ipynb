```python
import numpy as np
import pandas as pd
import cv2
import seaborn as sns
import matplotlib.pyplot as plt
# plt.axis("off")
```

```python
img = cv2.imread("mandelbrot-serial.ppm", cv2.IMREAD_GRAYSCALE)
```

```python
for thread in [2,3,4]:
    print([img[i*(len(img) // thread):(i+1)*(len(img) // thread)].sum() for i in range(thread)])

for thread in [2,3,4]:
    for i in range(thread):
        plt.figure()
        plt.axis("off")
        plt.imshow(img[i*(len(img) // thread):(i+1)*(len(img) // thread)], cmap='gray', vmin=0, vmax=255)
```

```python
for thread in [2,3,4]:
    print([img[i::thread].sum() for i in range(thread)])

for thread in [2,3,4]:
    for i in range(thread):
        plt.figure()
        plt.axis("off")
        plt.imshow(img[i::thread], cmap='gray', vmin=0, vmax=255)
```

```python
data = pd.DataFrame(
    np.array([
        [1,2,3,4],
        [.5, .3, .25, .2]
    ]).T,
    columns=['Thread Used', 'Runtime']
)
```

```python
sns.lineplot(
    data = data,
    x = 'Thread Used',
    y = 'Runtime'
)
```
